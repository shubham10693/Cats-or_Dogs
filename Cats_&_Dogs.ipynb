# Cats and Dogs Classification
Classifying whether a given image is a cat or a dog using Convolution Neural Network (CNN)

Dataset Link: https://www.kaggle.com/tongpython/cat-and-dog

```python
import keras
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
%matplotlib inline
from keras.callbacks import ModelCheckpoint
```

# Data Preprocessing

```python
train_datagen = ImageDataGenerator(width_shift_range=0.2, 
                                   height_shift_range=0.2,
                                   rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

train_dataset =  train_datagen.flow_from_directory('training_set',
                                                   batch_size=32,
                                                   target_size=(64, 64),
                                                   class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)

test_dataset = test_datagen.flow_from_directory('test_set',
                                                batch_size=32,
                                                target_size=(64, 64),
                                                class_mode='binary')
```

```python
train_dataset.class_indices
```

# Building CNN Model

```python
cnn_model =  keras.models.Sequential([
             keras.layers.Conv2D(filters=32, input_shape=[64,64,3], kernel_size= 3, activation='relu'),
             keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
             keras.layers.Conv2D(filters=64, kernel_size= 3, activation='relu'),
             keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
             keras.layers.Conv2D(filters=128, kernel_size= 3, activation='relu'),
             keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
    
             keras.layers.Dropout(0.5),
             keras.layers.Flatten(),
             keras.layers.Dense(units=128, activation='relu'), #Input Layer
             keras.layers.Dropout(0.2),
             keras.layers.Dense(units=256, activation = 'relu'),
             keras.layers.Dropout(0.3),
             keras.layers.Dense(units=1, activation='sigmoid')
])
```

```python
cnn_model.summary()
```

```python
cnn_model.compile( optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Training Dataset

```python
history = cnn_model.fit(train_dataset,
                        epochs=30,
                        verbose=1,
                        validation_data=test_dataset)
```

# Model Evaluation

```python
#Summarize the history of accuracy
plt.figure(figsize=(10,8))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epoch')
plt.ylabel('val_accuracy')
plt.title('Accuracy')
plt.legend(['train','test'], loc ='upper left')
plt.show()
```

```python
#Summarize the history of Loss
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('val_loss')
plt.title('Loss')
plt.legend(['train','test'], loc ='upper right')
plt.show()
```
